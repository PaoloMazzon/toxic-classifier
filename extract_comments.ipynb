```python
import pandas as pd
import random
```

```python
dfs = []
dfs.append(pd.read_csv("CONDA_train.csv"))
df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
trimmed_df = df[df['intentClass'].isin(['E', 'I'])].reset_index(drop=True)
trimmed_df.head()
```

```python
final_df = trimmed_df[['utterance']].copy()
final_df['utterance'] = final_df['utterance'].str.replace(r'\[SEPA\]', '', regex=True)
final_df.head()
```

```python
final_df['utterance'].to_csv('toxic_utterances.txt', index=False, header=False)
```
